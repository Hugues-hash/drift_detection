# 03 — Regime Detection

**Goal:** Identify distinct market regimes in the SPY return series using three complementary methods and characterise their statistical properties.

---

### Methods compared
| Method | Type | Temporal dependency | Key assumption |
|---|---|---|---|
| **HMM** (2-state) | Probabilistic | Yes — Markov chain | Returns are Gaussian per regime |
| **K-Means** (3-state) | Clustering | No | Regimes separable in vol-feature space |
| **CUSUM** | Sequential test | Yes — recursive | Mean shift in normalised returns / vol |

Using multiple methods guards against false discoveries that any single approach might produce.

```python
import sys
sys.path.insert(0, '../src')

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.patches as mpatches
import seaborn as sns

from preprocessing import preprocess_pipeline
from features import build_feature_matrix
from regime_detector import (
    HMMRegimeDetector,
    VolatilityRegimeDetector,
    detect_change_points,
    regime_statistics,
    REGIME_NAMES,
)
from rolling_stats import rolling_ks_test, cusum_detection

sns.set_theme(style='whitegrid', palette='muted', font_scale=1.05)
plt.rcParams.update({'figure.dpi': 120})

# ── Load data ──────────────────────────────────────────────────────────────
data        = preprocess_pipeline()
prices      = data['prices']
log_returns = data['log_returns']

print(f'Dataset: {len(log_returns):,} daily log returns  '
      f'({log_returns.index[0].date()} → {log_returns.index[-1].date()})')
```

---
## Method 1: Hidden Markov Model (2-State)

The **Gaussian HMM** models the return sequence as emissions from a latent Markov chain.  
Two hidden states are fitted:
- **State 0 — low volatility** (calm / bull market)  
- **State 1 — high volatility** (stress / bear market)

The Baum-Welch EM algorithm estimates transition probabilities and per-state Gaussian parameters simultaneously.

```python
hmm = HMMRegimeDetector(n_regimes=2, n_iter=1000, random_state=42)
hmm_regimes = hmm.fit_predict(log_returns)

print('HMM fit complete.')
print(f'  Regime 0 (low-vol)  : {(hmm_regimes == 0).sum():,} days  ({100*(hmm_regimes==0).mean():.1f}%)')
print(f'  Regime 1 (high-vol) : {(hmm_regimes == 1).sum():,} days  ({100*(hmm_regimes==1).mean():.1f}%)')
```

```python
# ── Emission statistics per regime ────────────────────────────────────────
print('\n── Emission Statistics ───────────────────────────────────────────────')
print(hmm.emission_stats.to_string())

print('\n── Transition Matrix ─────────────────────────────────────────────────')
print(hmm.transition_matrix.round(4).to_string())
print()
print('Interpretation: high diagonal values mean regimes are persistent.')
print('E.g. P(stay in low-vol | low-vol) should be close to 1 for a stable bull run.')
```

```python
# ── Regime statistics (returns within each state) ─────────────────────────
stats = regime_statistics(log_returns, hmm_regimes)
print('\n── Per-Regime Return Statistics ──────────────────────────────────────')
stats
```

```python
# ── Price chart with HMM regime shading ───────────────────────────────────
palette = {0: '#2ecc71', 1: '#e74c3c'}
regime_labels = {0: 'Low Volatility', 1: 'High Volatility'}

fig, axes = plt.subplots(2, 1, figsize=(15, 8), sharex=True)

# Panel 1: Price
axes[0].plot(prices.index, prices.values, color='#2c3e50', linewidth=0.9, zorder=2)
axes[0].set_ylabel('Price (USD)')
axes[0].set_title('SPY Price with HMM Regime Overlay')

# Panel 2: Returns
axes[1].bar(log_returns.index, log_returns.values,
            color=np.where(log_returns.values >= 0, '#aed6f1', '#f1948a'),
            width=1, alpha=0.6, zorder=1)
axes[1].set_ylabel('Log Return')
axes[1].set_title('Daily Log Returns with HMM Regime Overlay')
axes[1].axhline(0, color='black', linewidth=0.5)

# Shade regime periods on both panels
aligned = hmm_regimes.reindex(prices.index, method='ffill')
for ax in axes:
    current, start = None, None
    for date, regime in aligned.items():
        if regime != current:
            if current is not None:
                ax.axvspan(start, date, alpha=0.18,
                           color=palette[int(current)], lw=0)
            current, start = regime, date
    if current is not None:
        ax.axvspan(start, aligned.index[-1], alpha=0.18,
                   color=palette[int(current)], lw=0)

# Legend
patches = [mpatches.Patch(color=palette[k], alpha=0.5, label=regime_labels[k])
           for k in sorted(palette)]
axes[0].legend(handles=patches, loc='upper left', fontsize=9)

axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Annotate known crisis periods
for ax in axes:
    for dt, label in [('2008-09-15', 'GFC'), ('2020-03-11', 'COVID'), ('2022-01-03', '2022 Bear')]:
        ax.axvline(pd.Timestamp(dt), color='grey', linestyle=':', alpha=0.6, linewidth=0.8)

fig.tight_layout()
plt.show()
```

```python
# ── Return distribution per regime ────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))

for regime_id, color in palette.items():
    r = log_returns[hmm_regimes == regime_id]
    r.plot.hist(bins=80, alpha=0.55, color=color, ax=ax,
                label=f'{regime_labels[regime_id]} (n={len(r):,})', density=True)

ax.set_xlabel('Log Return')
ax.set_ylabel('Density')
ax.set_title('Return Distribution by HMM Regime')
ax.legend(fontsize=10)
fig.tight_layout()
plt.show()
```

---
## Method 2: K-Means Volatility Clustering (3-State)

K-Means groups trading days by their rolling volatility profile into 3 clusters:
- **Cluster 0** — low volatility (calm bull market)
- **Cluster 1** — medium volatility (uncertain / transitioning)
- **Cluster 2** — high volatility (crisis / bear market)

Unlike HMM, K-Means makes no temporal dependency assumption — each day is assigned independently based on its volatility features.

```python
km = VolatilityRegimeDetector(n_regimes=3, vol_window=20, random_state=42)
km_regimes = km.fit_predict(log_returns)

# Regime 2 maps to 'transition' in REGIME_NAMES
for k in range(3):
    count = (km_regimes == k).sum()
    name = {0: 'low-vol', 1: 'high-vol', 2: 'transition'}.get(k, f'regime_{k}')
    print(f'  Cluster {k} ({name:<12s}): {count:,} days  ({100*count/len(km_regimes):.1f}%)')
```

```python
km_stats = regime_statistics(log_returns, km_regimes)
print('\n── K-Means Per-Cluster Statistics ────────────────────────────────────')
km_stats
```

```python
# ── Price chart with K-Means regime shading ───────────────────────────────
km_palette = {0: '#2ecc71', 1: '#e74c3c', 2: '#f39c12'}
km_labels  = {0: 'Low Vol', 1: 'High Vol', 2: 'Transition'}

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(prices.index, prices.values, color='#2c3e50', linewidth=0.9, zorder=2)
ax.set_ylabel('Price (USD)')
ax.set_title('SPY Price with K-Means Volatility Regime Overlay (3 states)')

aligned_km = km_regimes.reindex(prices.index, method='ffill')
current, start = None, None
for date, regime in aligned_km.items():
    if regime != current:
        if current is not None:
            ax.axvspan(start, date, alpha=0.20,
                       color=km_palette[int(current)], lw=0)
        current, start = regime, date
if current is not None:
    ax.axvspan(start, aligned_km.index[-1], alpha=0.20,
               color=km_palette[int(current)], lw=0)

patches = [mpatches.Patch(color=km_palette[k], alpha=0.5, label=km_labels[k])
           for k in sorted(km_palette)]
ax.legend(handles=patches, loc='upper left', fontsize=9)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
fig.tight_layout()
plt.show()
```

---
## Method 3: CUSUM Change-Point Detection

CUSUM accumulates signed deviations from the historical mean.  
When the cumulative sum exceeds a decision threshold, a **structural break** is flagged.

We apply CUSUM to **rolling volatility** rather than raw returns — this makes it more sensitive to regime-level shifts in the variance process.

```python
# Apply CUSUM to rolling 20-day realised volatility (more informative than raw returns)
roll_vol_20d = log_returns.rolling(20).std().dropna() * np.sqrt(252)

cusum_df = cusum_detection(roll_vol_20d, threshold=4.0, drift_allowance=0.5)
change_pts = detect_change_points(roll_vol_20d, threshold=4.0, drift_allowance=0.5)

print(f'Number of CUSUM change points detected: {len(change_pts)}')
print('Dates:')
for cp in change_pts:
    print(f'  {cp.date()}')
```

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# Panel 1: Rolling vol
axes[0].plot(roll_vol_20d.index, roll_vol_20d.values,
             color='#e74c3c', linewidth=0.9, label='20-day Annualised Vol')
for cp in change_pts:
    axes[0].axvline(cp, color='#2c3e50', linestyle='--', alpha=0.7, linewidth=1.2)
axes[0].set_ylabel('Annualised Vol')
axes[0].set_title('Rolling 20-day Volatility with CUSUM Change Points')
axes[0].legend(fontsize=9)

# Panel 2: CUSUM statistics
axes[1].plot(cusum_df.index, cusum_df['cusum_pos'], color='#2ecc71',
             linewidth=0.9, label='CUSUM S⁺ (upper side)')
axes[1].plot(cusum_df.index, cusum_df['cusum_neg'], color='#e74c3c',
             linewidth=0.9, label='CUSUM S⁻ (lower side)')
axes[1].set_ylabel('CUSUM Statistic')
axes[1].set_title('CUSUM Statistics S⁺ and S⁻')
axes[1].legend(fontsize=9)

# Panel 3: Alarm
axes[2].fill_between(cusum_df.index, cusum_df['alarm'].astype(int),
                     step='mid', color='#e74c3c', alpha=0.7, label='Alarm (threshold exceeded)')
axes[2].set_ylabel('Alarm')
axes[2].set_yticks([0, 1])
axes[2].set_yticklabels(['No', 'Yes'])
axes[2].set_title('CUSUM Alarm Signal')
axes[2].legend(fontsize=9)
axes[2].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

fig.suptitle('CUSUM Change-Point Detection on Rolling Volatility', fontsize=13, fontweight='bold')
fig.tight_layout()
plt.show()
```

---
## Method 4: Rolling KS Test — Distribution Shift over Time

The two-sample KS test compares a rolling **current** window against a **reference** window.  
A rising KS statistic (and falling p-value below 0.05) indicates the return distribution has shifted.

```python
ks_df = rolling_ks_test(log_returns, ref_window=252, test_window=63)

drift_pct = 100 * ks_df['drift_flag'].mean()
print(f'Fraction of time with significant distribution shift (p<0.05): {drift_pct:.1f}%')

fig, axes = plt.subplots(2, 1, figsize=(14, 7), sharex=True)

axes[0].plot(ks_df.index, ks_df['ks_stat'], color='#2980b9', linewidth=1.0)
axes[0].set_ylabel('KS Statistic')
axes[0].set_title('Rolling KS Test Statistic (reference=252d, test=63d)')

axes[1].plot(ks_df.index, ks_df['ks_pvalue'], color='#8e44ad', linewidth=1.0, label='p-value')
axes[1].axhline(0.05, color='red', linestyle='--', linewidth=0.9, label='α = 0.05')
axes[1].fill_between(ks_df.index, ks_df['ks_pvalue'].values, 0.05,
                     where=(ks_df['ks_pvalue'].values < 0.05),
                     alpha=0.25, color='red', label='Significant drift')
axes[1].set_ylabel('p-value')
axes[1].set_title('KS Test p-value  (below red line = significant distribution shift)')
axes[1].legend(fontsize=9)
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

fig.suptitle('Rolling Kolmogorov-Smirnov Distribution Drift Test', fontsize=13, fontweight='bold')
fig.tight_layout()
plt.show()
```

---
## Comparison: HMM vs K-Means Regime Agreement

```python
# Align both regime series on common dates
common_idx = hmm_regimes.index.intersection(km_regimes.index)
hmm_aligned = hmm_regimes.reindex(common_idx)
km_aligned  = km_regimes.reindex(common_idx)

# Binary agreement: both say high-vol vs both say low-vol
# Treat K-Means cluster 0 as low-vol, clusters 1+2 as high-vol
km_binary = (km_aligned > 0).astype(int)

agreement = (hmm_aligned == km_binary).mean()
print(f'Binary regime agreement (HMM vs K-Means): {agreement:.1%}')

# Cross-tabulation
cross = pd.crosstab(
    hmm_aligned.rename('HMM Regime'),
    km_aligned.rename('K-Means Cluster'),
    margins=True
)
print('\nCross-tabulation (counts):')
print(cross.to_string())
```

---
## Summary

| Method | Key finding |
|---|---|
| **HMM** | Cleanly separates calm (low-vol) and stress (high-vol) regimes; transition probabilities confirm high regime persistence |
| **K-Means** | Identifies a third 'transition' state between calm and crisis, adding nuance |
| **CUSUM** | Detects structural breaks at major market events (GFC 2008, COVID 2020, etc.) |
| **Rolling KS** | Distribution drift is episodic — concentrated around crisis periods — not continuous |
